##Create Streamlit App File

```python
%%writefile /content/app.py
import streamlit as st
import torch
from PIL import Image
from diffusers import StableDiffusionPipeline, StableDiffusionImg2ImgPipeline

# ✅ Set Background Color (Light Yellow) & Text Color (Black)
st.markdown("""
    <style>
        .stApp {
            background-color: #fffacd; /* Light Yellow */
            color: #000000; /* Black Text */
        }
        h1, h2, h3, h4, h5, h6, p, label {
            color: #000000 !important; /* Ensure all text is black */
        }
    </style>
""", unsafe_allow_html=True)

# ✅ Check CUDA Availability
device = "cuda" if torch.cuda.is_available() else "cpu"
if device == "cpu":
    st.warning("⚠️ CUDA not available, using CPU. This will be slower.")

# ✅ Load Models
@st.cache_resource
def load_models():
    text_to_image_pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
    ).to(device)

    image_to_image_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
    ).to(device)

    return text_to_image_pipe, image_to_image_pipe

text_to_image_pipe, image_to_image_pipe = load_models()

# ✅ Streamlit UI
st.title("FIXIT: Virtual Fashion Designer")
st.write("Add a little tweak to your design based on your text or image input!")

# ---- 🎨 Text-to-Image Section ----
st.header("Text-to-Image")
text_prompt = st.text_area("Describe your outfit idea:", placeholder="A red satin evening gown with lace sleeves", height=150)

if text_prompt:
    with st.spinner("Generating outfit from text..."):
        try:
            generator = torch.Generator(device).manual_seed(0)
            text_image = text_to_image_pipe(prompt=text_prompt, generator=generator).images[0]
            st.image(text_image, caption="Generated from Text", use_column_width=True)
        except Exception as e:
            st.error(f"❌ Error generating image: {e}")

# ---- 🖼️ Image-to-Image Section ----
st.header("Image-to-Image")
uploaded_image = st.file_uploader("Upload an image for reference:", type=["png", "jpg", "jpeg"])

if uploaded_image is not None:
    try:
        pil_image = Image.open(uploaded_image).convert("RGB")
        st.image(pil_image, caption="Uploaded Image", use_column_width=True)

        image_prompt = st.text_area(
            "Describe the desired modifications:",
            placeholder="Add floral patterns, change to blue",
            height=100
        )

        if image_prompt:
            with st.spinner("Generating modified outfit..."):
                try:
                    generator = torch.Generator(device).manual_seed(0)

                    # ✅ Resize image for compatibility
                    pil_image = pil_image.resize((512, 512))

                    modified_image = image_to_image_pipe(
                        prompt=image_prompt,
                        image=pil_image,
                        strength=0.75,
                        guidance_scale=7.5,
                        generator=generator
                    ).images[0]

                    st.image(modified_image, caption="Modified Image", use_column_width=True)
                except Exception as e:
                    st.error(f"❌ Error generating modified image: {e}")
    except Exception as e:
        st.error(f"❌ Error loading image: {e}")
```
